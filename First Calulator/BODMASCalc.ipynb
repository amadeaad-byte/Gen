```python
#Creating a Function

#name your function and add amount of variables (keep note of the order)
def add_numbers(a,b):
    #what do you want the funtion to do with them?
    result = a + b
    #showing the result of the abive calculation
    return result

#testing out your function
print(add_numbers(5,3))
```

```python
#Creating a Function 2
def calc_profit(rev,costs):
    profit = rev - costs
    return profit

#testing out your function
print(calc_profit(1000,300))
```

```python
#Creating a Function 3 - Freestyle!
def money(current,salary,costs):
    bank_acc = (current + salary) - costs
    return bank_acc

print(money(1000,500,250))
```

```python
#Creating a Function - Alt Return Method
def avg_order_val(total_rev,order_amount):
    average = total_rev / order_amount
    return average

result = avg_order_val(1500,5)
print(result)
```
