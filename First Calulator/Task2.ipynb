You are a data analyst working for a fitness app, and you are given the following information: the total number of minutes users spent exercising today is 360, and the number of users is 12.
Your task is to write a Python function called "average_workout_time" that takes two inputs: total minutes and number of users.
 
Inside the function, you should calculate the average workout time by dividing the total minutes by the number of users, and then return the result.
After creating the function, call it using the values provided and print the output.
 
Finally, think about what the result tells you about user behaviour and how long, on average, each user is exercising per day.
 

---TASK 2---
You are a data analyst working for a delivery company, and you are given the following information: the total distance traveled by drivers today is 500 km, and the number of deliveries completed is 25.
Your task is to write a Python function called "average_distance_per_delivery" that takes two inputs: total distance and number of deliveries. Inside the function, you should calculate the average distance per delivery by dividing the total distance by the number of deliveries, and then return the result.
After creating the function, call it using the values provided and print the output.
Finally, think about what the result tells you about delivery efficiency and how far, on average, each delivery requires a driver to travel.

```python
#Task 1
def average_workout_time(total_mins, user_amount):
    workout = total_mins / user_amount
    return workout

print(average_workout_time(360,12))
```

```python
#Task 2
def average_distance_per_delivery(distance,deliveries):
    avg_dist = distance / deliveries
    return avg_dist

print(average_distance_per_delivery(500,25))
```
