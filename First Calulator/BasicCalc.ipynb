```python
# Addition

product_a = 240
product_b = 60
total_price = product_a + product_b

print(total_price)
```

```python
# Total Sum

sales = [300, 134, 68, 380]
total_rev = sum(sales)

print(total_rev)
```

```python
# Subtraction

revenue = 1000
refund = 200
net_rev = revenue - refund

print(net_rev)
```

```python
#Subracting in Lists

rev = 1000
refunds = [50, 30, 30]
net_r = sum(refunds)

print(net_r)
```

```python
#Multiplication

orders = 50
value_per = 20
total_sales = orders * value_per

print(total_sales)
```

```python
#Multiplying in Lists

orders = [2, 4, 6, 8]
price_per = 50
total = sum(orders) * price_per

print(total)
```

```python
#Division

total_money = 1000
amount = 25
amount_per = total_money / amount

print(amount_per)
```

```python
# Averages (keep in mind so you don't have to do them yourself!)

sales = [200, 150, 250, 300, 250, 300, 175]

# this is the command you use to find the average of a set of numbers
avg_sales = sum(sales) / len(sales)

print (avg_sales)
```

```python
# Finding the Sum in a Dictionary
customer_sales = {
    "Alice": 200,
    "bob": 150,
    "diana": 300,
    "charlie": 250
}

#Use a "." and "()" to do this!
total_rev = sum(customer_sales.values())

print(total_rev)
```
