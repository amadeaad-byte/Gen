# Instructions  
Write a program that takes two input numbers, then adds them together and prints the result.
then:
* Take two numbers and subtract the second from the first number.
* Take two numbers and multiply the two.
* Take two numbers and divide the first number by the second number.
* Take 3 numbers and add them together.

```python
# Part 1 - Addition
print("input first number")
num1 = int(input())
print("input second number")
num2 = int(input())
total = num1 + num2

print(total)

# Note: typing "int" before an input brackets tells Python that the content inside will be an integer (aka a whole number).
# if we don't use it, Python will treat the content as a string (aka text)
```

```python
#Part 2 - Subtrat
print("input first number")
num1 = int(input())
print("input second number")
num2 = int(input())
total = num1 - num2

print(total)
```

```python
# Part 3 - Multiplication
print("input first number")
num1 = int(input())
print("input second number")
num2 = int(input())
total = num1 * num2

print(total)
```

```python
# Part 4 - Division
print("input first number")
num1 = int(input())
print("input second number")
num2 = int(input())
total = num2 / num1

print(total)
```

```python
# Part 5 - Adding Multiple Numbers
num_list = [2, 4, 5, 8, 10]
total = sum(num_list)

print(total)
```
